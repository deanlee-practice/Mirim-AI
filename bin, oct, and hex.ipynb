- range(start, stop, step): start부터 stop-1까지, step만큼 증가하는 수열

```python
range(10)
```

```python
list(range(10))
```

```python
set(range(10))
```

```python
def my_fn(a, b, c):
  return a+b+c, a*b*c
```

```python
def my_fn1(a, b, c):
  return (a+b+c, a*b*c)
```

```python
my_fn(1, 2, 3)
```

```python
my_fn1(1, 2, 3)
```

```python
nums = my_fn1(1, 2, 3)
print(nums[0])
print(nums[1])
```

```python
(a, b) = my_fn1(1, 2, 3)
print(a)
print(b)
```

```python
a, b = my_fn1(1, 2, 3)
print(a)
print(b)
```

```python
(a, _, (_, c)) = (1, 2, (3, 4))
print(a)
print(c)
```

```python
(a, _, (_, (c, d))) = (1, 2, (3, (4, 5)))
```

```python
a, c, d
```
