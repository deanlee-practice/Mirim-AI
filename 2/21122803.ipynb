# 조건문 축약표현
- 3항 연산자 대안
- VALUE1 if **[CONDITION]** else VALUE2

```python
x = 1231
```

```python
"even" if x % 2 == 0 else "odd"
```

```python
result = "even" if x % 2 == 0 else "odd"
result
```

```python
while True:
  value = int(input("Value: "))
  result = "even" if value % 2 == 0 else "odd"
  print(result)
```

# 리스트 축약표현 (List Comprehension)
- 리스트 축약표현
- 문자열 
- 중첩 축약표현

```python
my_list = []
for i in range(10):
  my_list.append(2*i + 1)
my_list
```

```python
[(2*i + 1) for i in range(10)]
```

```python
[i for i in range(20)]
```

```python
[i for i in range(20) if i % 2 == 1]
```

```python
my_odd = [(2*i+1) for i in range(10)]
my_odd
```

```python
[2*n for n in my_odd]
```

```python
def fn(value):
  return value + 1
```

```python
[fn(n) for n in my_odd]
```

```python
[(n, fn(n)) for n in my_odd]
```

```python
eng = [99, 75, 35, 83, 97, 100]
math = [67, 78, 100, 82, 85, 68]
```

```python
for pair in zip(eng, math):
  print(pair[0], pair[1]) # eng, math
```

```python
for e, m in zip(eng, math):
  print(f"eng={e}, math={m}")
```

```python
for pair in zip(eng*2, math):
  print(pair)
```

```python
[True if e > 70 and m > 70 else False for e, m in zip(eng, math)]
```

```python
my_subjects = ["kor", "math", "eng"]
my_scores = [95, 25, 30]
```

```python
for subject, score in zip(my_subjects, my_scores):
  print(subject, score)
```

```python
my_score = {}
for subject, score in zip(my_subjects, my_scores):
  my_score[subject] = score
my_score
```

```python
{subject: score for subject, score in zip(my_subjects, my_scores)}
```

```python
eng = [99, 75, 35, 83, 97, 100]
math = [67, 78, 100, 82, 85, 68]
```

```python
for e, m in zip(eng*2, math):
  print(e, m)
```

```python
for e, m in zip(eng*2, math, strict=True):
  print(e, m)
```

```python
for e1, m1, e2, m2 in zip(eng, math, eng, math):
  print(e1, m1, e2, m2)
```

```python
from itertools import zip_longest
```

```python
for e, m in zip_longest(eng*2, math, fillvalue=0):
  print(e, m)
```
