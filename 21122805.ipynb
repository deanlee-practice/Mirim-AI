# Counter

```python
my_list = list(range(10))
my_list
```

```python
my_counter = collections.Counter(my_list)
my_counter
```

```python
my_list2 = [1, 1, 1, 2, 2, 3]
my_counter2 = collections.Counter(my_list2)
my_counter2
```

```python
my_counter.update({1: 1})
my_counter
```

```python
my_counter.update([1, 1, 1])
my_counter
```

```python
my_counter.subtract({1: 1})
my_counter
```

```python
set(my_counter.elements())
```

```python
set(my_counter2.elements())
```

```python
my_counter.most_common()[0][0]
```

```python
my_counter.most_common()[0][1]
```

```python
my_counter.most_common()[0]
```

```python
my_counter.most_common(3)
```

```python
my_counter.most_common(1)[0]
```

```python
my_str = "aaabbbcc"
my_str
```

```python
my_counter = {}
for c in my_str:
  my_counter[c] = my_counter.get(c, 0) + 1
my_counter
```

```python
my_num = 222222
```

```python
my_counter = {}
for digit in bin(my_num)[2:]:
  my_counter[digit] = my_counter.get(digit, 0) + 1
my_counter
```

```python
my_counter['1']
```

```python
collections.Counter(bin(my_num)[2:])['1']
```

```python
my_counter = collections.Counter(bin(my_num)[2:])
my_counter
```

```python
my_counter['1']
```

```python
my_counter = {}
for c in f"{my_num:b}":
  my_counter[c] = my_counter.get(c, 0) + 1
my_counter
```

```python
my_counter = collections.Counter(f"{my_num:b}")
my_counter
```
