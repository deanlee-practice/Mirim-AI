## 정리
- 반환값이 없는 함수: return None
- 반환값이 있는 함수: return 값
- 반환값이 여러개인 함수: return 튜플

````python
res, _ = my_fn3(5, 6)
res
````

````python
def my_fn4(a, b, c, d):
  return a, (b, (c, (d)))
````

````python
res = my_fn4(1, 2, 3, 4)
res
````

````python
_, (a, (_, (b))) = my_fn4(1, 2, 3, 4)
print(a)
print(b)
````

## 가변 매개변수
- 0개 이상의 인자를 받을 수 있는 매개변수
- 위치 가변 매개변수
- 키워드 가변 매개변수 (이름이 있는 가변 매개변수)

````python
# 위치 가변 매개변수
# *args : unpacking (X)
def my_fn9(num, *args):
  print(args)
````

````python
my_fn9(1, 2, 3, 4)
````

````python
my_fn9(1)
````

````python
my_fn9(1, 2, 3, 4, 5, 6, 7, 8, 9)
````

````python
def my_fn10(num, *nums):
  for n in nums:
    num += n
  return num
````

````python
my_fn10(1)
````

````python
my_fn10(1, 2)
````

````python
my_fn10(1, 2, 3)
````

````python
def my_fn11(num, *nums):
  if nums:
    return num + sum(nums)
  else:
    return num
````

# 매개변수 펼치기(unfold / unpacking)
- 리스트/튜플 펼치기
- dict 펼치기

````python
my_scores1 = [92, 88, 76]
my_scores1
````

````python
my_scores2 = {
    "a": 92,
    "b": 88,
    "c": 92
}
my_scores2
````

````python
# 위치 매개변수 / 이름이 있는 매개변수 (keyword)
def my_fn16(a, b, c):
  print(a)
  print(b)
  print(c)
````

````python
my_fn16(1, 2, 3)
````

````python
my_fn16(b=2, a=1, c=3)
````

````python
my_fn16(my_scores1[0], my_scores1[1], my_scores1[2])
````

````python
my_fn16(a=my_scores2['a'], b=my_scores2['b'], c=my_scores2['c'])
````

````python
my_fn16(*my_scores1) # == my_fn16(my_scores1[0], my_scores1[1], my_scores1[2]) # 리스트, 튜플
````

````python
my_fn16(**my_scores2) # == my_fn16(a=my_scores2['a'], b=my_scores2['b'], c=my_scores2['c'])
````

````python
3 * 4
````

````python
my_scores3 = [92, 88, 76, 100]
````

````python
my_fn16(*my_scores3) # == my_fn16(92, 88, 76, 100)
````

````python
my_scores4 = {
    'a': 92,
    'b': 88,
    'z': 76
}
````

````python
my_fn16(**my_scores4) # == my_fn16(a=92, b=88, z=76) == my_fn16(key1=value1, key2=value2, key3=value3)
````

# 일급 클래스 (First Class) 함수
- 변수: 변수에 저장 가능
- 함수: 매개변수로 전달 가능
- 함수: 반환값으로 사용 가능
- 함수는 객체의 레퍼런스다.

````python
def my_fn20():
  return "Hello"
````

````python
my_fn20()
````

````python
a = my_fn20()
a
````

````python
b = my_fn20
b
````

````python
b() # == my_fn20()
````

````python
def my_fn21(a):
  print(a)
  value = a() # == my_fn20()
  print(value)
````

````python
my_fn21(my_fn20)
````

````python
def my_fn22():
  def my_fn23():
    return "Hello"
  return my_fn23
````

````python
my_fn23()
````

````python
a = my_fn22()
a
````

````python
a() # == my_fn23
````

````python
def my_fn24(a):
  print(f"a={a}")
  return a
````

````python
my_fn24(my_fn20)
````

````python
c = my_fn24(my_fn20)
c()
````
