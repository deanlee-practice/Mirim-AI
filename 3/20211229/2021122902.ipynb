# 문자열의 포매팅
- **{:[[fill]align][sign][#][0][width][,][.precision][type]}**

## 포메팅
- 서식: "나이: ___, 이름: ___"
- 데이터의 형태

## 파이썬이 다루는 데이터
- 문자열
- 숫자
  - 정수
  - 실수
  - 2진수, 8진수, 16진수
    - 5진수, 1234나라 (1, 2, 3, 4, 11, 12, 13, 14, 21, 

### 서식
- % 연산자 (Python2)
- format 메소드 (Python3)
- fstring (강추)

```python
"%c %i" % (97, 97)
```

```python
"국어점수: {}, 영어점수: {}".format(96, 100)
```

```python
"국어점수: {0}, 영어점수: {0}".format(96, 100, 97)
```

```python
my_list = [96, 100]
"국어점수: {0[0]}, 영어점수: {0[1]}".format(my_list)
```

```python
kor = 96
eng = 100
```

```python
f"국어점수: {kor} 영어점수: {eng}"
```

## 데이터의 표현
- **{:[[fill]align][sign][#][0][width][,][.precision][type]}**

- fill: 혼자 사용 불가능, 채워지는 문자
- align: 정렬
  - 오른쪽 정렬: >
  - 왼쪽 정렬: <
  - 가운데 정렬: ^
- sign:
  - 기본값: 마이너스(-) 부호만 표시
  - +: 양수도 부호 표시 (26 => +26)
  - -: (기본값), 마이너스(-) 부호만 표시
  - '공백': 양수의 + 부호를 공백으로 표시 (26 => ' 26')
- #: 2진수, 8진수, 16진수 각각에 '0b', '0o', '0x'
- 0: 0으로 나머지 공간을 채움
- comma(,): 천의 단위로 comma 표시
- width: 사용할 공간, 문자열을 자르지 않는다. (width가 3일때, Python)
- .precision: 소숫점 이하의 갯수
- type: 데이터의 타입
  - b(2진수), o(8진수), x(16진수), d(정수), f(실수), c(문자), s(문자열), %(소수->백분률)

- 문자열:
  - fill, align, width
- 숫자:
  - 정수:
    - sign, comma(,), 0, width
  - 실수:
    - sign, .precision, 0, width
  - n-ary(2진수, 8진수, 16진어)
    - sign, #, 0, width
