# Set의 연산

- 교집합: and (&)
- 합집합: or (|)
- 차집합: -
- 대칭차집합: XOR(^)

image.png

```python
my_basket = {'apple', 'banana', 'orange', 'pear'}
my_basket2 = {'apple', 'banana', 'grape'}
```

```python
True and True
```

```python
False or False
```

```python
my_basket & my_basket2
```

```python
my_basket | my_basket2
```

```python
my_basket - my_basket2
```

```python
my_basket2 - my_basket
```

```python
(my_basket | my_basket2) - (my_basket & my_basket2)
```

```python
my_basket2 ^ my_basket # (XOR)
```

```python
my_basket.intersection(my_basket2) # == my_basket | my_basket2
```

```python
my_basket2.intersection(my_basket) # == my_basket2 | my_basket
```

```python
my_basket.union(my_basket2)
```

```python
my_basket.difference(my_basket2) # == my_basket - my_basket2
```

```python
my_basket2.difference(my_basket) # == my_basket2 - my_basket
```

```python
my_basket2.symmetric_difference(my_basket) # my_basket ^ my_basket2
```
