# 브로드캐스트(Broadcast)
- 두 배열의 차원(shape)이 같지 않을 때:
- 낮은 차원 -> 높은 차원

image.png

```python
np.array([[0, 2, 4], [6, 8, 10]]) + np.array([100, 200, 300])
```

```python
np.array([[0, 2, 4], [6, 8, 10]]) + np.array([[100, 200, 300], [100, 200, 300]])
```

```python
np.array([[0, 2, 4], [6, 8, 10]]) + np.array([[100,], 
                                              [200,]])
```

```python
np.array([[0, 2, 4], [6, 8, 10]]) + np.array([[100, 100, 100],
                                              [200, 200, 200]])
```

```python
np.array([[0, 2, 4], [6, 8, 10]]) + 3
```

```python
np.array([[0, 2, 4], [6, 8, 10]]) + np.array([[3, 3, 3], [3, 3, 3]])
```

```python
my_array = np.array([[5, 2, 3],
                     [4, 3, 1]])
my_array
```

```python
# axis = -1 가장 큰 축
np.sort(my_array, axis=-1)
```

```python
np.sort(my_array, axis=1)
```

```python
my_array
```

```python
# [A, B]
# [5, 2, 3] =>A, [4, 3, 1] => B
np.sort(my_array, axis=0)
```

```python
# 5, 2, 3
# 4, 3, 1
np.sort(my_array, axis=1)
```
