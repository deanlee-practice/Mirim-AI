```python
np.concatenate([[1,2], [2, 3], [4,5]])
```

```python
# np.concatenate([[A, B], [C, D]], axis=0) => [A, B, C, D]
# [[1, 2], [3, 4]] + [[5, 6], [7, 8]]
np.concatenate([[[1, 2], [3, 4]], [[5, 6], [7, 8]]], axis=0)
```

```python
# [[1, 2] + [5, 6], [3, 4]+[7, 8]]
np.concatenate([[[1, 2], [3, 4]], [[5, 6], [7, 8]]], axis=1)
```
