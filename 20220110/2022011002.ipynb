# NumPy 데이터 타입
- 정수
- 실수
- 복소수(Complex)
- 부울

# 정수(Integer)
- 부호가 존재
  - numpy.int
  - numpy.int8
  - numpy.int16
  - numpy.int32
  - numpy.int64

- 부호가 없음
  - numpy.uint8
  - numpy.uint16
  - numpy.uint32
  - numpy.uint64

```python
import numpy as np
```

```python
np.int(32.22) # numpy.int(32.22) == int(32.22)
```

```python
np.int32(32.22)
```

```python
np.uint64(32.22)
```

```python
isinstance(int(32.22), np.int)
```

```python
isinstance(int(32.22), np.uint64)
```

실수 (Float, Real Number)
- numpy.float
- numpy.float16
- numpy.float32
- numpy.float64

```python
np.float(32) # == float(32)
```

```python
np.float(32.22)
```

```python
np.float64(32.22)
```

# 복소수 (Complex)
- np.complex == complex
- np.complex64
- np.complex128
- np.complex256

```python
2+3j
```

```python
np.complex(2, 3)
```

```python
np.complex64(2+3j)
```

```python
np.complex64(2)
```

```python
isinstance(2+3j, np.complex)
```

# 부울 (Boolean)
- 0, 1
- True / False (Python)
- np.bool
- np.bool8

```python
np.bool(0), np.bool(1), np.bool(2)
```

```python
np.bool8(0), np.bool8(1), np.bool8(2)
```
