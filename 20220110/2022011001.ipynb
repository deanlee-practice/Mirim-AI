```python
my_privates = """
  홍길동 900101-1011111
  고길동 801231-2022222
  이둘리 990326-1234567
  김둘리 011211-3234567
  박군인 800905-1049118
  김선생 070417-4059119
  김구라 070417-5059119
  박구라 070417-6059119
"""
```

```python
result = []
for private in my_privates.strip().split("\n"):
  splited_private = private.strip().split(" ")
  result.append(" ".join([splited_private[0], splited_private[1][:8] + "*" * 7]))
result
```

```python
my_pat = re.compile("\d{6}-\d{7}")
my_pat.findall(my_privates)
```

```python
my_pat = re.compile("\d{6}-[1|2|3|4]\d{6}")
my_pat.findall(my_privates)
```

```python
my_pat1 = re.compile("(\d{6}-[1|2|3|4])\d{6}")
my_pat1.findall(my_privates)
```

```python
for mat in my_pat1.finditer(my_privates):
  print(mat.group())
  print(mat.group(1))
```
