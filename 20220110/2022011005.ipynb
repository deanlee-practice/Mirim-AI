# 임의 숫자 생성

- numpy.random.rand: 0과 1 사이의 난수 생성

  - numpy.random.randf
  - numpy.random.random_sample
  - numpy.random.sample
- numpy.random.randint: 특정 범위를 가지는 정수인 임의 숫자 생성
  - numpy.random.random_integer
- numpy.random.randn: 표준 정규 분포에서 표집, 평균 0, 표준편차 1
  - numpy.random.standard_normal:
- numpy.random.normal: 정규 분포(m, s)에서 표집 
- numpy.random.choice: 배열에서 숫자 n개 선택
- numpy.random.bytes
- numpy.random.seed:

```python
my_nums = np.random.randn(10000)
```

```python
fig, axes = plt.subplots()
axes.hist(my_nums, bins=np.linspace(-5, 5, 100))
plt.show()
```

```python
np.random.normal(0, 1, 5) # == np.random.randn(5)
```

```python
my_nums = np.random.normal(0, 1, 10000)
```

```python
fig, axes = plt.subplots()
axes.hist(my_nums, bins=np.linspace(-5, 5, 100))
plt.show()
```

```python
my_nums = np.random.normal(0, 1, 100000)
my_nums1 = np.random.normal(1, 1, 100000)
my_nums2 = np.random.normal(0, 1.3**2, 100000)
fig, axes = plt.subplots()
axes.hist(my_nums, bins=np.linspace(-5, 5, 100), label='$N(0, 1^2)$', alpha=0.3)
axes.axvline(0, 0, 1, color='blue', alpha=0.3)
axes.hist(my_nums1, bins=np.linspace(-5, 5, 100), color='red', label='$N(1, 1^2)$', alpha=0.3)
axes.axvline(1, 0, 1, color='red', alpha=0.3)
axes.hist(my_nums2, bins=np.linspace(-5, 5, 100), color='green', label='$N(0, 1.3^2)$', alpha=0.3)
axes.axvline(0, 0, 1, color='green', alpha=0.3)
axes.legend()
plt.show()
```
