# 축약형 (Comprehension)
- 조건 축약표현 (condition comprehension)
- 리스트/튜플 축약표현(list/tuple comprehension)
- dic 축약표현 (dict comprehension)

```python
value = 7
True if value % 2 == 1 else False
```

```python
[i for i in range(20) if i % 2 == 1]
```

```python
print([True if i % 2 == 1 else False for i in range(20)])
```

```python
my_list = []
for i in range(20):
  if i % 2 == 1:
    my_list.append(True)
  else:
    my_list.append(False)
```

```python
def my_fn(n):
  if n%2==1:
    return True
  else:
    return False
```

```python
print(list(map(my_fn, range(20))))
```

```python
print(list(map(lambda n: True if n %2==1 else False, range(20))))
```

```python
print(tuple((True if n % 2 == 1 else False for n in range(20))))
```

```python
[True if i % 2 == 1 else False for i in range(20)]
```

```python
(True if n % 2 == 1 else False for n in range(20))
```

# dict 축현 표현
- 키와 값이 분리되어 있을 때
- 키와 값이 원하는 형식이 아닐 때
- 키와 값을 바꾸고 싶을 때

```python
my_subjects = ["kor", "math", "eng"]
my_scores = [95, 25, 30]
```

```python
# 1. key: subject, value: score
{subject: score for subject, score in zip(my_subjects, my_scores)}
```

```python
my_dict = {
    "a": [92, 87, 66],
    "b": [82, 82, 99],
    "c": [100, 97, 77]
}
```

```python
# 2. key: key, value: mean(평균)
{k: sum(v)/len(v) for k, v in my_dict.items()}
```

```python
# 3. key: 대문자, value: value
{k.upper(): v for k, v in my_dict.items()}
```

```python
my_dict1 = {
    "a": 10,
    "b": 20,
    "c": 30
}
```

```python
# 4. key: value, value: key
{v:k for k, v in my_dict1.items()}
```
