## 반환값
- 반환값이 없는 함수: return None
- 반환값이 있는 함수: 
  - 반환값이 여러개인 함수: return 튜플

## 매개변수의 정의
- 위치 매개 변수 (positional parameter / argument)
- 기본 매개 변수 (default parameter / argument)
- 이름이 있는 매개변수 (키워드 매개변수)
- 가변 매개변수: 
  - 가변 위치 매개 변수
  - 가변 키워드 매개 변수

````python
# from functools import reduce
import functools
````

````python
my_list = list(range(5))
my_list
````

````python
list(map(lambda x: x**2, my_list))
````

````python
# x    , y => return
# 0    , 1 => 0 + 1
# 0 + 1, 2 => 1 + 2
# 1 + 2, 3 => 3 + 3
# ...
functools.reduce(lambda x, y: x+y, my_list)
````

````python
# x      , y => return
# 100    , 0 => 100
# 100    , 1 => 101
# 101    , 2 => 103
# ...
functools.reduce(lambda x, y: x+y, my_list, 100)
````

````python
pow(2, 3)
````

````python
# 2 ** 3 % 2
pow(2, 3, 2)
````

````python
my_fn = functools.partial(pow, 2)
````

````python
my_fn(3) # == pow(2, 3)
````

````python
my_fn(4) # == pow(2, 4)
````

````python
my_fn1 = functools.partial(pow, 2, 3)
````

````python
my_fn1(2) # == my_fn1(2, 3, 2)
````

````python
my_fn1(3) # == my_fn1(2, 3, 3)
````

````python
my_fn2 = lambda x: pow(2, x)
````

````python
my_fn2(3) # == pow(2, 3)
````

````python
my_fn2(4) # == pow(2, 4)
````

````python
my_fn3 = lambda x: pow(x, 2)
````

````python
my_fn3(2) # == pow(2, 2)
````

````python
my_fn3(3) # == pow(3, 2)
````
