# 이터러블(Iterable)의 최대최소


```python
# dict + iterable
blackpink = [
  {
    "name": "지수",
    "age": 25,
    "height": 162
  },
  {
    "name": "제니",
    "age": 24,
    "height": 163
  },
  {
    "name": "로제",
    "age": 23,
    "height": 168
  },
  {
    "name": "리사",
    "age": 24,
    "height": 166.5
  },
]
```

```python
def get_age(elem):
  return elem['age']
```

```python
max(blackpink, key=get_age)
```

```python
max(blackpink, key=lambda m:m['age'])
```

```python
min(blackpink, key=lambda m:m['age'])
```

```python
max(blackpink, key=lambda m:m['height'])
```

```python
min(blackpink, key=lambda m:m['height'])
```
